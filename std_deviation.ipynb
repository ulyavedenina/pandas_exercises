```python
import pandas as pd
import numpy as np; np.random.seed(42)
import matplotlib.pyplot as plt
from matplotlib.transforms import Affine2D
```

```python
df = pd.read_csv("languages.tsv", sep='\t')
```

```python
# This code iterates over results of LAS (Labeled Attachment Score) and UAS (Unabeled Attachment Score)
# for 8 Tupí-Guaraní languages, outputting their standard deviation


x = ('Akuntsu', 'Guajajara', 'Mbya', 'Karo', 'Munduruku', 'Nheengatu', 'Tupinamba', 'Teko', 'English' )

uas_cols = [col for col in df.columns if '_UAS' in col]
las_cols = [col for col in df.columns if '_LAS' in col]

uas_std = [np.std(df[x].tolist()) for x in uas_cols]
las_std = [np.std(df[x].tolist()) for x in las_cols]

uas_mean = [np.mean(df[x].tolist()) for x in uas_cols]
las_mean = [np.mean(df[x].tolist()) for x in las_cols]

fig, ax = plt.subplots(figsize=(10, 6))

val1 = Affine2D().translate(-0.1, 0.0) + ax.transData
val2 = Affine2D().translate(+0.1, 0.0) + ax.transData

print(las_mean)
print(las_std)

er1 = ax.errorbar(x, uas_mean, yerr=uas_std, marker="o", linestyle="none", transform=val1)
er2 = ax.errorbar(x, las_mean, yerr=las_std, marker="o", linestyle="none", transform=val2)

ax.legend([er1, er2], ['UAS', 'LAS'], loc='upper right')

plt.show()
```
